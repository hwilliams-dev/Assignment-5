```python
def visualize(filename):
    """Returns visualization outputs for the AVG, MAX, MIN results for each imported dataset."""
   
    # import the data and convert to a dataframe: 'data'
    data = pd.read_csv(filepath_or_buffer = filename)
   
    # calculate the Avg, Max, Min values
    avg_result = data.mean()
    max_result = data.max()
    min_result = data.min()
   
    # Plot the line graphs for AVG, MAX, MIN
    avg_result.plot(subplots=True, label = "Avg", figsize = (9,7), color = 'deeppink', linewidth = 2)
    max_result.plot(subplots=True, label = "Max", figsize = (9,7), color = 'slateblue', linewidth = 2)
    min_result.plot(subplots=True, label = "Min", figsize = (9,7), color = 'lightseagreen', linewidth = 2)
  
    # Plot the legend, x & y labels and tick marks
    plt.title("Dataset AVG, MAX & MIN", fontsize= 16)
    plt.legend(loc='center right')
    plt.xlabel('Feature#', fontsize = 13, fontweight = 'bold')
    plt.ylabel('Values', fontsize = 13, fontweight = 'bold')
    plt.xticks(fontsize = 11)
    plt.yticks(fontsize = 11)
    plt.show()
    
def detect_problems(filename):
    """Detects integrity issues contained in imported datasets."""
    
    # import the data and convert to a dataframe: 'data'
    data = pd.read_csv(filepath_or_buffer = filename)
    
    # calculate the Max, Min and Sum of Min values
    max_res = data.max()
    min_res = data.min()
    sum_res = min_res.sum()
    
    # Check Max and Sum_Res values for integrity
    if data['0'].max() == 0 and data['20'].max() == 20: # trigger data values 0 & 20 == 0 doubts integrity
        print("\nI doubt the integrity of this dataset.\n\n")
    else:    
        if sum_res == 0: # checks to see if sum result == 0 and questions why it is happening
            print("Why is this happening?\n\n")
        else:
            print("Dataset looks good.\n\n") # if none of the conditions above are met, data is clear
        
 
  # Import appropriate libraries
import glob
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

dataset = sorted(glob.glob('data*.csv'))
  
```

```python
for datasets in dataset[:4]:
    print(datasets)
    visualize(datasets)
    detect_problems(datasets)
```
